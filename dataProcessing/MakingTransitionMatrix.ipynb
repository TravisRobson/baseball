```python
## I need a way to translate the codes in the PBP to rows of balls and strikes

## Codes and assumptions
    ## IN -- in play so the state moves to the next batter-pitcher with [0,0]
    ## S  -- strike -- easy
    ## B   -- ball --easy
    ## F   -- foul ball on pitchout (what happens?)
    ## FT  -- foul tip on bunt (what happens?)
    ## FB  -- foul bunt (so what? )
    ## HB  -- hit batter (I guess this just resets the strike ball states)
    ## IB  -- intentional ball (this is just a ball I think)
    ## MB  -- misses bunt attempt (isn't that a strike?)
    ## NP  -- no pitch (what?)
    ## PO  -- what
    ## POS -- swinging on pitchout (?)
    ## N-A -- unknown missed pitch damn
    
    ## the unique entries in this dataset are

def pbpToBallStrikeArray(pbp):
    ## the array is just going to be a pitchX2 array with each row being [#strike, #ball]
    sequence = pbp.split('/')
    
    arr = np.zeros(shape = (len(sequence), 2))
    
    balls   = 0
    strikes = 0
    
    for i, seq in enumerate(sequence):
        if seq in ['IN', 'HB']:
            
            break # new batter
        
        if seq in ['B'] :
            
            balls += 1
            arr[i + 1] = np.array([balls, strikes])
            
        if seq in ['S'] :
            
            strikes += 1
            arr[i + 1] = np.array([balls, strikes])
            
        if seq in ['F','FB','FT'] :
            ## foul -- either nothing happens or you 
            ## get a strike if strike < 2
            
            arr[i + 1] = np.array([balls, strikes])
            
            if strikes < 2:
                strikes += 1
                
            
    return arr
```

```python
## for a transition matrix the sum along a row should be 1 so that means that P_i,j (row, col) represents
## the probability of state going from state i -> j 

## Note that i am using transition vector super weirdly and wrongly, but its a good way to use it because
## I started calling it that in my head informally

def transitionVectorToCountMatrix(transitionVector):
    
    countMatrix = np.zeros(shape = (28, 28))
    
    for i, entryI in enumerate(transitionVector[:-1]):
        
        entryJ = transitionVector[i + 1]
        
        countMatrix[entryI][entryJ] += 1
        
    return countMatrix
```

```python
countMat = transitionVectorToCountMatrix(stateSeries)
```

```python
def countMatrixToTransitionMatrix( countMatrix ):
    
    transMat = deepcopy(countMatrix)
    
    for i, row in enumerate(transMat):
        
        sumRow = np.sum(row)
        
        if sumRow != 0:
            transMat[i] *= 1./sumRow
        
    return transMat
```

```python
transMat = countMatrixToTransitionMatrix(countMat)
```
